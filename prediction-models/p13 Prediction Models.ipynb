## Construcci√≥n de un modelo para predecir la cancelaci√≥n de usuarios

Aqu√≠ vamos a crear un modelo de clasificaci√≥n binaria para predecir si un cliente cancelar√° o no en el pr√≥ximo mes. Se utilizar√°n 2 modelos: el de regresi√≥n log√≠stica y el de bosque aleatorio, para comparar los resultados y concluir qu√© modelo dar√° mejores resultados

```python
# Definir las caracter√≠sticas y la variable objetivo 
X = df.drop('churn', axis=1) 
y = df['churn'] 

# Dividir los datos en entrenamiento y validaci√≥n (80/20)
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2)
```

### Modelo regresi√≥n logistica

```python
# Crear el modelo de regresi√≥n log√≠stica
log_model = LogisticRegression(max_iter=1000)

# Entrenar el modelo
log_model.fit(X_train, y_train)

# Realizar predicciones
y_pred_log = log_model.predict(X_valid)

# Evaluar el modelo
print(f'Modelo:{log_model}')
print(f'Exactitud: {accuracy_score(y_valid, y_pred_log)}')
print(f'Precisi√≥n: {precision_score(y_valid, y_pred_log)}')
print(f'Recall: {recall_score(y_valid, y_pred_log)}')
```

La exactitud del modelo predijo correctamente el 93.75% de las veces. En este caso, el modelo tiene una precisi√≥n del 87.38%, lo que significa que de todas las predicciones de cancelaci√≥n, aproximadamente el 87.38% fueron correctas. Un recall de 0.89 significa que el modelo identific√≥ correctamente al 89% de los clientes que cancelaron.

Las metricas de este modelo muestran que funciona bien, sin embargo hay que compararlo con el modelo de Random Forrest para detectar el de mejor desempe√±o.

### Modelo Bosque Aleatorio

```python
# Crear el modelo de bosque aleatorio
rf_model = RandomForestClassifier()

# Entrenar el modelo
rf_model.fit(X_train, y_train)

# Realizar predicciones
y_pred_rf = rf_model.predict(X_valid)

# Evaluar el modelo
print(f'Modelo:{rf_model}')
print(f'Exactitud: {accuracy_score(y_valid, y_pred_rf)}')
print(f'Precisi√≥n: {precision_score(y_valid, y_pred_rf)}')
print(f'Recall: {recall_score(y_valid, y_pred_rf)}')
```

El modelo de bosque aleatorio tiene una exactitud de 92.13%, lo que indica que es muy efectivo en t√©rminos generales, aunque es menor al de regresi√≥n log√≠stica. La precisi√≥n es muy similar a la de la regresi√≥n log√≠stica, sin embargo el recall es m√°s bajo que en el modelo de regresi√≥n log√≠stica, lo que sugiere que este modelo presenta m√°s falsos positivos.

### Comparaci√≥n de modelos

El modelo de regresi√≥n log√≠stica tiene un mejor desempe√±o en t√©rminos de recall, lo cual es clave para detectar correctamente a los clientes que van a cancelar, mientras que el modelo de bosque aleatorio tiene una mayor precisi√≥n y un recall ligeramente menor, lo que significa que es m√°s preciso cuando predice que un cliente cancelar√°, pero pierde m√°s clientes potenciales que realmente cancelan. 

Ambos modelos pueden funcionar dependiendo del objetivo, para minimizar la p√©rdida de clientes (identificar a los clientes que probablemente van a cancelar), la regresi√≥n ser√≠a la mejor opcion, ya que tiene un recall m√°s alto.
Por otro lado, para predecir correctamente a los clientes que cancelan y mantener una buena precisi√≥n, entonces el modelo de bosque ser√≠a la mejor opci√≥n.

## Crear cl√∫sters de usuarios

El objetivo de este paso es segmentar a los usuarios en diferentes grupos para analizar sus comportamientos y caracter√≠sticas. Se llevar√° a cabo la segmentaci√≥n con el algoritmo de KMeans. Primero se deben estandarizar los datos y despu√©s crear una matriz para aplicarle el algoritmo y lograr segmentar a los clientes. 

```python
#Estandarizar los datos
features_to_cluster = ['age', 'lifetime', 'avg_class_frequency_total', 'avg_class_frequency_current_month', 'avg_additional_charges_total']
X = df[features_to_cluster].dropna()  # Elimina valores faltantes

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Usamos StandardScaler para normalizar las caracter√≠sticas seleccionadas para que tengan una media de 0 y una desviaci√≥n est√°ndar de 1. Esto asegura que las caracter√≠sticas no sesguen el clustering.

```python
#Crear el dendrograma
linked = linkage(X_scaled, method='ward')  # Calcular la matriz de distancias

plt.figure(figsize=(10, 7))
dendrogram(linked)

# L√≠nea de corte: Puedes ajustar la altura de corte seg√∫n lo que necesites
plt.axhline(y=45, color='r', linestyle='--')  # La l√≠nea de corte roja a la altura 30 (ajustar seg√∫n tus necesidades)

plt.title('Dendrograma de Cl√∫steres')
plt.xlabel('√çndice de Clientes')
plt.ylabel('Distancia')
plt.show()
```

La funci√≥n linkage() genera la matriz de distancias entre los puntos (clientes), y con dendrogram() la visualizamos. Esto nos ayuda a decidir cu√°ntos cl√∫steres usar en el siguiente paso. Se traaz√≥ una l√≠nea de corte para contar los cl√∫steres, y se decidi√≥ que se agrupar√°n en 5 clusters.

```python
# Aplicar K-Means decidimos que el n√∫mero de cl√∫steres es 5
kmeans = KMeans(n_clusters=5, random_state=42)
df['cluster'] = kmeans.fit_predict(X_scaled)
```

El algoritmo de K-Means segmenta los clientes en 5 cl√∫steres, despu√©s de aplicar el algoritmo, asignamos a cada cliente un cl√∫ster usando kmeans.fit_predict()

```python
# Analizar los cl√∫steres y ver las caracter√≠sticas promedio de cada cl√∫ster
cluster_means = df.groupby('cluster')[features_to_cluster].mean()
display(cluster_means)
```

En esta tabla podemos ver las caracter√≠sticas promedio de cada grupo, podemos ver que el grupo 2 tiene un lifetime de aprox 12 meses, mientras que el grupo 0 tiene un lifetime de a penas 1 mes. Estas caracter√≠sticas nos ayudan a entender c√≥mo est√°n agrupados los datos. A continuaci√≥n se muestran las visualizaciones para las distintas caracter√≠sticas.

```python
# Visualizaci√≥n: distribuciones de caracter√≠sticas para cada cl√∫ster
for feature in features_to_cluster:
    plt.figure(figsize=(6, 4))
    sns.histplot(data=df, x=feature, hue='cluster', multiple="stack", kde=True)
    plt.title(f'Distribuci√≥n de {feature} por Cl√∫ster')
    plt.show()
```

```python
# Analizar la tasa de cancelaci√≥n por cl√∫ster
churn_rate_by_cluster = df.groupby('cluster')['churn'].mean()
print(churn_rate_by_cluster)
```

Se analiza la media de las caracter√≠sticas en cada cl√∫ster y se observan distribuciones para entender mejor c√≥mo se agrupan los usuarios. Tambi√©n calculamos la tasa de cancelaci√≥n por cl√∫ster para entender la propensidad de cancelaci√≥n de cada grupo, se puede observar que el cluster 0 tiene mayor propensidad a cancelar.
