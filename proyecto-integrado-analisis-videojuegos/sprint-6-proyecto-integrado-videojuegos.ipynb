## Limpieza de datos

En el apartado anterior, podemos ver que los t√≠tulos de las columnas de nuestros datos est√°n en un formato de may√∫sculas y min√∫sculas. Para hacer m√°s facil su manipulaci√≥n se convertir√°n todos los t√≠tulos a min√∫sculas.

```python
#cambiar columnas a minusculas
games_df.columns = games_df.columns.str.lower()
games_df.info()
```

Ahora, se corregir√°n los datos que tienen tipos incorrectos, en este caso, year_of_release a numero entero y user_score a n√∫mero, ya que es un objeto. Se utiliza 'coerce' en user_score ya que el tipo de dato era un objeto y necesitamos un valor num√©rico. Los valores que tenian el texto NaN ahora son valores nulos. En year_of_release se ignoraron los errores ya que no tenemos los datos correctos de los a√±os, y rellenar con 0 o con otro a√±o puede afectar el comportamiento de los datos. Los dem√°s datos tienen formato correcto.

```python
#cambiar tipos de datos
games_df['user_score'] = pd.to_numeric(games_df['user_score'], errors='coerce')
games_df['year_of_release']= games_df['year_of_release'].astype('int', errors='ignore')

games_df.info()
```

Como ya se vio anteriormente, no hay datos duplicados, sin embargo, si hay datos ausentes en algunas columnas:

```python
#encontrar valores ausentes
print('El n√∫mero de ausentes es:')
print(games_df.isna().sum() )
```

Se observa que no hay datos ausentes en cuanto a las plataformas o en cuanto a las ventas, es por esto que todas las filas de nuestros datos contienen informaci√≥n relevante. Para las columnas de datos object (name, genre y rating), podemos rellenar con la palabra 'Unkown', ya que nos ayuda a reducir los valores ausentes sin eliminar parte de las ventas.

```python
#limpiar columnas
columns_to_clean = ['name' , 'genre' , 'rating']

for column in columns_to_clean:
    games_df[column] = games_df[column].fillna('unknown')
```

Ahora, se elegir√°n las plataformas con m√°s ventas hist√≥ricas y se agrupar√°n loa juegos por a√±o para ver c√≥mo se comparan las ventas a lo largo de los a√±os. Esto es para saber cu√°nto tiempo dura la popularidad de las diferentes plataformas.

```python
top_platforms = sales_by_platform.head(15).index

df_top = games_df[games_df['platform'].isin(top_platforms)]

sales_by_year = df_top.groupby(['year_of_release', 'platform'])['total_sales'].sum().unstack()

sales_by_year.plot(kind='line', figsize=(10, 6))
plt.title('Distribuci√≥n de ventas por a√±o (plataformas m√°s exitosas)')
plt.ylabel('Ventas (millones)')
plt.xlabel('A√±o')
plt.show()
```

Con la gr√°fica anterior podemos ver que las ventas de las plataformas suben r√°pidamente en su lanzamiento pero a los pocos a√±os bajan y se vuelven menos populares hasta despaarecer. De cualquier forma, se tiene que hacer un esudio m√°s exacto para que podamos saber cu√°ntos a√±os de datos son relevantes para el an√°lisis.
A continuaci√≥n, se calcula el promedio de a√±os que una plataforma se mantiene activa en cuanto a las ventas totales. Para esto, primero se visualizan plataformas que antes ten√≠an ventas pero que ya no. Se utiliz√≥ el a√±o 2000 como par√°metro de filtrado, 15 a√±os antes de los datos presentados.

```python
# Encontrar plataformas con ventas en el pasado
past_sales = games_df.groupby('platform')['year_of_release'].max()

# Ver plataformas que no tienen ventas recientes (por ejemplo, despu√©s de 2000)
vintage_platforms = past_sales[past_sales < 2000]
print('Plataformas que sol√≠an ser populares pero ya no tienen ventas:')
print(vintage_platforms)
```

Ahora se calcula el tiempo de vida de las plataformas. Para esto, calculamos el primer y ultimo a√±o en el que una plataforma tuvo ventas, despu√©s con estos datos agregamos una columna llamada 'lifespan' o tiempo de vida, en la que se calculan los a√±os activos. Despu√©s se toma un promedio de estos a√±os para tener el dato de cu√°ntos a√±os duran las ventas de una plataforma en promedio.

```python
# Calcular el primer y √∫ltimo a√±o de ventas para cada plataforma
sales_active_years = games_df.groupby('platform')['year_of_release'].agg(['min', 'max'])

# A√±adir una columna de la duraci√≥n de vida de cada plataforma
sales_active_years['lifespan'] = sales_active_years['max'] - sales_active_years['min']
print(sales_active_years)

# Promedio del tiempo que tardan en desaparecer las plataformas
avg_lifespan = sales_active_years['lifespan'].mean()
print(f'El promedio de duraci√≥n de las plataformas es de {avg_lifespan} a√±os.')
```

En promedio, el tiempo de vida activo de una plataforma es de 7.61 a√±os. Bas√°ndonos en esto, hay que tomar en cuenta lo siguiente:

Considerar plataformas activas en los √∫ltimos 7 a√±os: Estas plataformas a√∫n pueden ser relevantes y tener potencial en el a√±o entrante.
Identificar tendencias de plataformas emergentes: Revisar las plataformas que han aparecido en los √∫ltimos 1-3 a√±os, ya que suelen tener un crecimiento r√°pido si est√°n ganando popularidad.
Descartar plataformas que han desaparecido o mostrado declive: Si una plataforma no ha tenido ventas en los √∫ltimos 2-3 a√±os, probablemente ya no sea relevante para el pr√≥ximo a√±o.

```python
#Filtrar los juegos tomando en cuenta el promedio de duracion de las plataformas
current_year = 2016
recent_games_df = games_df[games_df['year_of_release'] >= (current_year - (avg_lifespan))]
```

### An√°lisis de crecimiento 

A continuaci√≥n, se mostrar√°n las plataformas con mayor crecimiento de ventas en los √∫ltimos 3 a√±os, junto con lo que representan estos √∫ltimos 3 a√±os de ventas:

```python
# Agrupar por plataforma y calcular el crecimiento en los √∫ltimos 3 a√±os
growing_games_df = recent_games_df[recent_games_df['year_of_release'] >= (current_year - 3)]
growing_platforms = growing_games_df.groupby('platform')['total_sales'].sum().sort_values(ascending=False)

# Ver plataformas con mayor crecimiento en los √∫ltimos 3 a√±os
print('Plataformas con mayor crecimiento en los √∫ltimos 3 a√±os:')
display(growing_platforms)

growing_platforms.plot(kind='bar', figsize=(10, 6))
plt.title('Plataformas con mayor crecimiento en los √∫ltimos 3 a√±os:')
plt.ylabel('Ventas (millones)')
plt.xlabel('Consolas')
plt.show()
```

A continuaci√≥n se calculan las plataformas nuevas,  y sus ventas en los ultimos a√±os:

```python
new_platforms = recent_games_df[recent_games_df['year_of_release'] >= (current_year - 2)].groupby('platform')['total_sales'].sum().sort_values(ascending = False)
print('Plataformas nuevas con sus ventas en los √∫ltimos 2 a√±os:')
print(new_platforms)

new_platforms.plot(kind='bar', figsize=(10, 6))
plt.title('Plataformas nuevas y sus ventas en los √∫ltimos 2 a√±os:')
plt.ylabel('Ventas (millones)')
plt.xlabel('Consolas')
plt.show()
```

A continuaci√≥n buscamos las plataformas que lanzaron juegos hace m√°s tiempo, podemos ver su √∫ltimo a√±o de lanzamiento lo que nos da informaci√≥n sobre su uso:

```python
declining_platforms = recent_games_df.groupby('platform')['year_of_release'].max()
declining_platforms = declining_platforms[declining_platforms < (current_year - 1)].sort_values()
print('Plataformas que podr√≠an desaparecer y su √∫ltimo a√±o de lanzamiento:')
print(declining_platforms)

```

Ahora se har√° un diagrama de caja para las ventas totales de todos los juegos, por plataforma, para los juegos recientes. 

```python
# Crear el diagrama de caja
plt.figure(figsize=(12, 8))
sns.boxplot(x='platform', y='total_sales', data=recent_games_df)

# T√≠tulo y etiquetas
plt.title('Diagrama de caja de las ventas globales por plataforma')
plt.xlabel('Plataformas')
plt.ylabel('Ventas Globales (millones)')
plt.yscale('log')
plt.xticks(rotation=45) 
plt.show()
```

En el diagrama anterior podemos ver que por los datos at√≠picos es dificil observar el diagrama para cada plataforma. Esto puede ser por que hay juegos con ventas muy por encima que otros.
```python
# Agrupar por plataforma y calcular la media y la mediana de las ventas globales
avg_sales = recent_games_df.groupby('platform')['total_sales'].agg(['mean', 'median']).sort_values(by='mean', ascending=False)

# Mostrar los resultados
print(avg_sales)
```

Estos valores at√≠picos de juegos con ventas excepcionales hacen que las ventas promedio se eleven con respecto al resto de los datos, en la lista de medias y medianas de las ventas totales, podemos observar que en casi todas las plataformas, la media est√° bastante por encima de la mediana. Por ejemplo, en el caso de Wii, se observa que hay un juego con m√°s de 30 millones en ventas, sin embargo la mayor√≠a de los datos que vemos dentro de los bigotes est√° muy por debajo de los 5 millones en ventas. ## Pruebas de Hip√≥tesis

Las hip√≥tesis que se busca probar en este proyecto son:
- [ ] Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son las mismas.
- [ ] Las calificaciones promedio de los usuarios para los g√©neros de Acci√≥n y Deportes son diferentes.

Para ambas se redactar√° una hipotesis nula y una hip√≥tesis alternativa y se elaborar√° una prueba de hip√≥tesis de una cola, ya que se est√° comparando la media de dos poblaciones en ambos casos. La redeacci√≥n de las hip√≥tesis se elaborar√° siendo la hip√≥tesis nula que no hay diferencia entre las medias de las poblaciones y la hip√≥tesis alternativa que s√≠ hay una diferencia signigicativa.

Para probar las hip√≥tesis se realizar√° una prueba estad√≠stica de t de student para encontrar si la diferencia entre las medias de las dos poblaciones es significativa. Se establecer√° el valor de alpha en 5%.

### Primera hip√≥tesis: diferencias en las calificaciones de Xbox ONe y PC

Hip√≥tesis nula: Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son las mismas.
Hip√≥tesis alternativa: Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC diferentes.

```python
df_xbox = recent_games_df[recent_games_df['platform'].str.contains('XOne')]
df_pc = recent_games_df[recent_games_df['platform'].str.contains('PC')]

#eliminar valores nulos en user_score
df_xbox = df_xbox.dropna(subset=['user_score'])
df_pc = df_pc.dropna(subset=['user_score'])

#comprobar diferencia de varianza
var_xbox = df_xbox['user_score'].var()
var_pc = df_pc['user_score'].var()

print(f"Varianza para Xbox One: {var_xbox}")
print(f"Varianza para PC: {var_pc}")
```

```python
alpha = 0.05

results = st.ttest_ind(df_xbox['user_score'] , df_pc['user_score'], equal_var = False)

print('valor p:', results.pvalue)

if ( results.pvalue < alpha) :
    print("Rechazamos la hip√≥tesis nula")
else:
    print("No podemos rechazar la hip√≥tesis nula")
```

Obtuvimos como resultado que no podemos rechazar la hip√≥tesis nula, por lo que no hay una diferencia significativa entre las calificaciones de los usuarios para los juegos en xbox one y pc.

### Segunda hip√≥tesis: diferencias en las calificaciones para Acci√≥n y Deportes

Hip√≥tesis nula:Las calificaciones promedio de los usuarios para los g√©neros de Acci√≥n y Deportes son iguales.
Hip√≥tesis alternativa: Las calificaciones promedio de los usuarios para los g√©neros de Acci√≥n y Deportes son diferentes.


```python
df_action = recent_games_df[recent_games_df['genre'].str.contains('Action')]
df_sports = recent_games_df[recent_games_df['genre'].str.contains('Sports')]

#eliminar valores ausentes
df_action = df_action.dropna(subset=['user_score'])
df_sports = df_sports.dropna(subset=['user_score'])

#valores de varianza
var_action = df_action['user_score'].var()
var_sports = df_sports['user_score'].var()

print(f"Varianza para action: {var_action}")
print(f"Varianza para sports: {var_sports}")
```

```python
alpha = 0.05

results = st.ttest_ind(df_action['user_score'] , df_sports['user_score'], equal_var = False)

print('valor p:', results.pvalue)

if ( results.pvalue < alpha) :
    print("Rechazamos la hip√≥tesis nula")
else:
    print("No podemos rechazar la hip√≥tesis nula")
```

Obtuvimos como resultado el rechazo de la hip√≥tesis nula, por lo que podemos decir que hay una diferencia significativa en la calificacion de los usuarios para los g√©neros de acci√≥n y deportes.
