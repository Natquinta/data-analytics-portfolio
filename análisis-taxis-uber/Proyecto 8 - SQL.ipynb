### Planteamiento de hipótesis



Hipótesis nula (H₀): La duración promedio de los viajes desde el Loop al Aeropuerto Internacional O'Hare no cambia los sábados lluviosos.

Hipótesis alternativa (H₁): La duración promedio de los viajes desde el Loop al Aeropuerto Internacional O'Hare cambia los sábados lluviosos.

Para realizar un análisis estadístico se dividirán los datos en dos grupos, con buen clima o mal clima. Se determinará si las varianzas son iguales con una prueba de Levene en la que se asignará un valor alpha de 0.5. Después se realizara una prueba T para comprobar si se rechaza la hipótesis nula.

```python
# Dividir los datos en dos grupos
bad_weather = df_trips_weather[df_trips_weather['weather_conditions'] == 'Bad']['duration_seconds']
good_weather = df_trips_weather[df_trips_weather['weather_conditions'] == 'Good']['duration_seconds']

# Realizar la prueba de Levene para verificar si las varianzas son iguales
levene_stat, levene_pvalue = stats.levene(bad_weather, good_weather)

# Imprimir el resultado de la prueba de Levene
print("Prueba de Levene :", levene_stat)
print("Valor p de Levene:", levene_pvalue)

# Si el valor p de la prueba de Levene es menor que el nivel de significación (por ejemplo, 0.05), 
# las varianzas no son iguales y usamos `equal_var=False` en la prueba t.

alpha = 0.05
if levene_pvalue < alpha:
    print("Las varianzas son diferentes. Usaremos equal_var=False en la prueba t.")
    t_stat, p_value = stats.ttest_ind(bad_weather, good_weather, equal_var=False)
else:
    print("Las varianzas son iguales. Usaremos equal_var=True en la prueba t.")
    t_stat, p_value = stats.ttest_ind(bad_weather, good_weather, equal_var=True)
```

```python
# Resultados de la prueba t
print("Estadístico t:", t_stat)
print("Valor p:", p_value)

# Interpretar el valor p
if p_value < alpha:
    print("Rechazamos la hipótesis nula: La duración de los viajes cambia los sábados lluviosos.")
else:
    print("No rechazamos la hipótesis nula: No hay evidencia suficiente para afirmar que la duración de los viajes cambia los sábados lluviosos.")
```
