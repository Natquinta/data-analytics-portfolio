### Visitas

A continuaci√≥n se har√° un c√°lculo de las m√©tricas y trazo de gr√°ficos de las visitas al sitio para contestar las siguientes preguntas: 
- ¬øCu√°ntas personas lo usan cada d√≠a, semana y mes?
- ¬øCon qu√© frecuencia los usuarios regresan?
- ¬øCu√°ntas sesiones hay por d√≠a? (Un usuario puede tener m√°s de una sesi√≥n).
- ¬øCu√°l es la duraci√≥n de cada sesi√≥n?


```python
#C√°lculo de visitas por d√≠a semana y mes

visits['session_year'] = visits['Start Ts'].dt.isocalendar().year

visits['session_month'] = visits['Start Ts'].dt.month

visits['session_week'] = visits['Start Ts'].dt.isocalendar().week

visits['session_date'] = visits['Start Ts'].dt.date


dau = visits.groupby(['session_date']).agg({'Uid':'nunique'}).mean()

print('Daily active users average:',int(dau))

wau= visits.groupby(['session_year','session_week']).agg({'Uid': 'nunique'}).mean()

print('Weekly active users average:',int(wau))

mau= visits.groupby(['session_year','session_month']).agg({'Uid':'nunique'}).mean()

print('Monthly active users average:',int(mau))

sticky_wau = dau / wau * 100
print('Sticky wau:', round(sticky_wau), '%')
sticky_mau = dau / mau * 100
print('Sticky mau:',round(sticky_mau), '%')
```

En promedio, hay 907 usuarios activos cada d√≠a, y 23,228 usuarios activos por semana y por mes. 
El factor de adherencia semanal es del 16% y el mensual es de 4%. 
Esto nos indica que 16% de los usuarios semanales est√°n activos diariamente en la plataforma, mientras que 4% de los usuarios activos mensualmente est√°n activos diariamente. 
Al ser una empresa de venta de entradas para eventos, no representa una m√©trica muy significativa, ya que puede haber usuarios que s√≥lamente la usen cuando hay un evento que les interese.

```python
#Cuantas sesiones hay por d√≠a

daily_visits = visits.groupby(['session_date'])['Uid'].count().sort_values()
print(daily_visits)
daily_visits_avg = daily_visits.mean()
print()
print('Promedio de sesiones al d√≠a: ',daily_visits_avg)
```

Las visitas diarias tienen un promedio de 987 por d√≠a, esto es tomando en cuenta que un usuario puede acceder varias veces al sitio. Se ordenaron los valores para ver cual es el numero m√°s bajo de visitas al d√≠a, se observa que el 31 de marzo de 2018 solamente hay un registro, y el d√≠a con m√°s visitas fue el 24 de noviembre 2017, con 4,042 visitas.

```python
sessions_per_user = visits.groupby(['session_year', 'session_month']).agg({'Uid': ['count','nunique']})

sessions_per_user.columns=['n_sessions','n_users']

sessions_per_user['sess_per_user']= sessions_per_user['n_sessions'] / sessions_per_user['n_users']

print(sessions_per_user)
print()
print('Visitas mensuales promedio por usuario: ', sessions_per_user['sess_per_user'].mean())
```

En promedio, los clientes visitan 1.28 veces el sitio de forma mensual.

```python
 #Duraci√≥n de cada sesi√≥n
visits['session_duration'] = (
    visits['Start Ts'] - visits['End Ts']
).dt.seconds

visits['session_duration'].hist(bins=100)
```

En el gr√°fico, se puede observar que una gran cantidad de sesiones terminan muy pronto, y el resto no se distribuya de manera uniforme. Es considerablemente m√°s frecuente una sesi√≥n m√°s larga. Es por esto que se debe utilizar la moda para calcular la duraci√≥n promedio de una sesi√≥n en lugar de la media.


```python
asl = visits['session_duration'].mode()
print('Average Session Length (seconds): ', asl)
```

### Ventas

Ventas:

- ¬øCu√°ndo empieza la gente a comprar? (En el an√°lisis de KPI, generalmente nos interesa saber el tiempo que transcurre entre el registro y la conversi√≥n, es decir, cuando el usuario se convierte en cliente. Por ejemplo, si el registro y la primera compra ocurren el mismo d√≠a, el usuario podr√≠a caer en la categor√≠a Conversion 0d. Si la primera compra ocurre al d√≠a siguiente, ser√° Conversion 1d. Puedes usar cualquier enfoque que te permita comparar las conversiones de diferentes cohortes para que puedas determinar qu√© cohorte o canal de marketing es m√°s efectivo.)
- ¬øCu√°ntos pedidos hacen durante un per√≠odo de tiempo dado?
- ¬øCu√°l es el tama√±o promedio de compra?
- ¬øCu√°nto dinero traen? (LTV)

Para analizar las conversiones se debe dividir los datos en cohortes para poder comparar entre s√≠. Se agrupar√°n por mes de registro las visitas y luego se unir√°n los datos de las ordenes para crear cohortes mensuales.Despu√©s se filtrar√° por los usuarios que hicieron una compra y se calcular√° la tasa de conversi√≥n por cohorte dividiendo el numero de usuarios que compraron entre el numero de usuarios en el cohorte.

```python
visits['Start Ts'] = pd.to_datetime(visits['Start Ts'])

visits['first_visit_month'] = visits.groupby('Uid')['Start Ts'].transform('min').dt.to_period('M')

orders['Buy Ts'] = pd.to_datetime(orders['Buy Ts'])
merged_data = visits[['Uid', 'first_visit_month']].merge(orders, on='Uid', how='left')

merged_data.head()
```

```python
# Usuarios √∫nicos por cohorte
users_per_cohort = merged_data.groupby('first_visit_month')['Uid'].nunique()

# Usuarios que compraron por cohorte
buyers_per_cohort = merged_data[~merged_data['Buy Ts'].isna()].groupby('first_visit_month')['Uid'].nunique()

# Tasa de conversi√≥n
conversion_rate_per_cohort = (buyers_per_cohort / users_per_cohort) * 100

print(conversion_rate_per_cohort)
```

```python
#gr√°fico conversion rate
conversion_rate_per_cohort.plot(kind='bar', figsize=(10, 6), title="Tasa de Conversi√≥n por Cohorte")
plt.ylabel('Tasa de Conversi√≥n (%)')
plt.xlabel('Cohorte (Mes/A√±o)')
plt.show()
```

La tasa de conversi√≥n por cohorte tuvo su valor m√°s alto en el grupo de junio 2017, mientras que el m√°s bajo fue en abril 2018.

```python
# Ingresos totales por cohorte
revenue_per_cohort = merged_data.groupby('first_visit_month')['Revenue'].sum()
print('Ingresos totales por cohorte: ', revenue_per_cohort)

# LTV por cohorte
ltv_per_cohort = revenue_per_cohort / users_per_cohort
print()
print('LTV por cohorte: ', ltv_per_cohort)
```

```python
revenue_per_cohort.plot(kind='line', figsize=(10, 6), marker='o', title="Ingresos por Cohorte")
plt.ylabel('Ingresos totales')
plt.xlabel('Cohorte (Mes/A√±o)')
plt.show()
```

```python
ltv_per_cohort.plot(kind='line', figsize=(10, 6), marker='o', title="LTV por Cohorte")
plt.ylabel('LTV (Ingresos por Usuario)')
plt.xlabel('Cohorte (Mes/A√±o)')
plt.show()
```

En las dos gr√°ficas anteriores, podemos ver que hay un pico en los ingresos totales y en el LTV en junio y en septiembre. Hay que hacer un an√°lisis m√°s profundo sobre qu√© acciones pudieron llevar a el aumento de ventas de entradas durante estos meses.

```python
# Crear columna para el mes/a√±o de la compra
merged_data['buy_month'] = merged_data['Buy Ts'].dt.to_period('M')

# Usuarios √∫nicos que compran por cohorte y mes de compra
cohort_retention = merged_data.pivot_table(
    index='first_visit_month', 
    columns='buy_month', 
    values='Uid', 
    aggfunc='nunique'
)
cohort_retention.head()
```

```python

plt.figure(figsize=(12, 8))
sns.heatmap(cohort_retention, annot=True, fmt='0')
plt.title("Retenci√≥n de Usuarios por Cohorte")
plt.ylabel('Cohorte (Mes/A√±o)')
plt.xlabel('Mes de Compra')
plt.show()
```

En el gr√°fico de mapa de calor se puede observar que la retenci√≥n del cohorte siempre es m√°s fuerte en el primer mes, es decir en el de registro, y de ah√≠ la retencion baja significativamente. Podr√≠a ser por el tipo de eventos que hay y la disponibilidad de boletos.

A continuaci√≥n se calcular√°n los costos totales, as√≠ como los costos por fuente de la venta y el costo por d√≠a

```python
total_costs = costs['costs'].sum()

costs_by_source = costs.groupby('source_id')['costs'].sum()

costs_over_time = costs.groupby('dt')['costs'].sum()
print('Costos totales: ', total_costs)
print()
print('Costos por canal: ', costs_by_source)
print()
print('Costos por d√≠a: ', costs_over_time)
```

```python
# Total de clientes por fuente
customers_by_source = visits.groupby('Source Id')['Uid'].nunique()

# CAC por fuente
cac_by_source = costs_by_source / customers_by_source

print(customers_by_source)
print()
print(cac_by_source)
```

```python
#Graficar CAC
cac_by_source_sorted = cac_by_source.sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(
    x=cac_by_source_sorted.index.astype(str), 
    y=cac_by_source_sorted.values, 
    palette="YlOrBr"
)

plt.title("Costo de Adquisici√≥n de Clientes (CAC) por Fuente", fontsize=16)
plt.xlabel("Fuente de Adquisici√≥n", fontsize=12)
plt.ylabel("CAC (Costo por Cliente)", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```

El gr√°fico nos permite observar que el costo por cliente m√°s alto es por mucho el 3, m√°s adelante se debe calcular el ROMI de √©ste para ver si la inversi√≥n ha resultado en una buena estrategia de mkt que aumente las ventas.Hay que combinar los datos de visitas, pedidos y costos de marketing para alinear los ingresos y los costos por fuente. Tambi√©n es importante analizar el CAC por cohorte para poder entender mejor los datos y el comportamiento del cliente.

```python
visits['Cohort'] = pd.to_datetime(visits['Start Ts']).dt.to_period('M')

customers_by_source_cohort = visits.groupby(['Source Id', 'Cohort'])['Uid'].nunique().reset_index()
customers_by_source_cohort.rename(columns={'Uid': 'Customers'}, inplace=True)

costs['Cohort'] = pd.to_datetime(costs['dt']).dt.to_period('M')
costs_by_source_cohort = costs.groupby(['source_id', 'Cohort'])['costs'].sum().reset_index()

cac_data = pd.merge(costs_by_source_cohort, customers_by_source_cohort, 
                    left_on=['source_id', 'Cohort'], 
                    right_on=['Source Id', 'Cohort'], 
                    how='left')

cac_data['CAC'] = cac_data['costs'] / cac_data['Customers']

cac_pivot = cac_data.pivot_table(
    index='Cohort', 
    columns='Source Id', 
    values='CAC'
)

display(cac_pivot)
```

```python

plt.figure(figsize=(12, 8))
sns.heatmap(cac_pivot, annot=True, fmt=".2f", cbar_kws={'label': 'Costo de Adquisici√≥n (CAC)'})
plt.title("Costo de Adquisici√≥n de Clientes (CAC) por Fuente y Cohorte", fontsize=16)
plt.xlabel("Fuente de Adquisici√≥n", fontsize=12)
plt.ylabel("Cohorte (Mes)", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
